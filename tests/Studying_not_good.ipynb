```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = '/Users/jimenagonzalez/research/DSPL/Simulations-Double-Source-Gravitational-Lensing/tests/'
filename = 'single_study.csv'
data = pd.read_csv(path+filename)
data['center'] = np.sqrt(data['c1x']**2 + data['c2y']**2)
not_good = data[data['i']==0]
good = data[data['i']==1]
regular = data[data['i']==3]
```

```python
not_good.head()
```

```python
plt.figure(figsize=(7,6))
plt.xlabel('Apparent Magnitude')
plt.ylabel('Einstein Radius')
plt.plot(not_good['m1'], not_good['theta_E'], 'bo', label='not good')
plt.plot(regular['m1'], regular['theta_E'], 'ko', label='regular')
plt.plot(good['m1'], good['theta_E'], 'ro', label='good')
plt.legend()
plt.show()
```

```python
x = np.linspace(0,5,10)
plt.figure(figsize=(7,6))
plt.xlabel('Center')
plt.ylabel('Einstein Radius')
plt.plot(not_good['center'], not_good['theta_E'], 'bo', label='not good')
plt.plot(regular['center'], regular['theta_E'], 'ko', label='regular')
plt.plot(good['center'], good['theta_E'], 'ro', label='good')
plt.plot(x,x,'g')
plt.xlim(0,1.4)
plt.ylim(0,2.3)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.xlabel('Redshift')
plt.ylabel('Apparent Magnitude')
plt.plot(not_good['z1'], not_good['m1'], 'bo', label='not good')
plt.plot(regular['z1'], regular['m1'], 'ko', label='regular')
plt.plot(good['z1'], good['m1'], 'ro', label='good')
plt.legend()
plt.show()
```
