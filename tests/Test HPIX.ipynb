```python
import healpy
import pandas as pd
import numpy as np
```

```python
path = '/Users/jimenagonzalez/research/DSPL/Simulations-Double-Source-Gravitational-Lensing/tests/'
filename = 'test_HPIX.csv'
data = pd.read_csv(path+filename)
data.head()
```

```python
PIXEL, PIXEL2, PIXEL3, NRA, NDEC = [], [], [], [], []
PIXEL4, PIXEL5, PIXEL6 = [], [], []
for index, row in data.iterrows():
    ra = row['RA']
    dec = row['DEC']
    pix = row['HPIX_4096']
    new_ra, new_dec = healpy.pix2ang(4096, pix, nest=True, lonlat=True)
    new_pix = healpy.ang2pix(4096, np.radians(-dec+90.), np.radians(360.-ra), nest=True)
    new_pix2 = healpy.ang2pix(4096, np.radians(-dec+90.), np.radians(360.-ra))
    new_pix4 = healpy.ang2pix(4096, 360.-ra, dec+90., lonlat = True, nest=True)
    new_pix5 = healpy.ang2pix(4096, 360.-ra, dec+90., lonlat = True)
    new_pix3 = healpy.ang2pix(4096, ra, dec, lonlat = True, nest = True)
    new_pix6 = healpy.ang2pix(4096, ra, dec, lonlat = True)
    NRA.append(new_ra)
    NDEC.append(new_dec)
    PIXEL.append(new_pix)
    PIXEL2.append(new_pix2)
    PIXEL3.append(new_pix3)
    PIXEL4.append(new_pix4)
    PIXEL5.append(new_pix5)
    PIXEL6.append(new_pix6)
```

```python
data.drop(columns=['TILENAME', 'COADD_OBJECT_ID', 'HPIX_16384'], inplace = True)
data['NRA'] = NRA
data['NDEC'] = NDEC
data['PIXEL1'] = PIXEL
data['PIXEL2'] = PIXEL2
data['PIXEL3'] = PIXEL3 #este es!!!
data['PIXEL4'] = PIXEL4
data['PIXEL5'] = PIXEL5
data['PIXEL6'] = PIXEL6
data.head()
```
