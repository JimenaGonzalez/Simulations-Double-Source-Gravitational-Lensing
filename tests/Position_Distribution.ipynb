```python
import numpy as np
import random
from scipy.stats import skewnorm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
r = skewnorm.rvs(a=0, size=10000, loc=0.0, scale=100)
plt.figure()
plt.hist(r)
plt.show()
```

```python
r = []
n = 0
for i in range(1000):
    t = skewnorm.rvs(a=-5, size=1000, loc=0.7, scale=0.6) #a: skew, loc: center, scale:sigma
    t = t[np.random.randint(0,1000)]
    r.append(t)
    if(t < 0.2 and t >-0.2): n +=1
print(n)
plt.figure()
plt.hist(r)
plt.show()
```

```python
#Distribution to use in my code
r = []
n = 0
for i in range(1000):
    t = skewnorm.rvs(a=-5, size=10000, loc=0.75, scale=0.425) #a: skew, loc: center, scale:sigma
    t = t[np.random.randint(0,10000)]
    s = 1 if random.randint(0, 1) else -1
    t = t*s
    r.append(t)
    if(t > 0.75): n +=1
print(n)
plt.figure()
plt.hist(r)
plt.show()
```
