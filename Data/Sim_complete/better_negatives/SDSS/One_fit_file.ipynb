```python
import numpy as np
import pandas as pd
import os
from tqdm.notebook import tqdm

from astropy.io import fits
from astropy.table import Table
```

```python
data = pd.read_csv('morphological_catalog.csv')
data.head()
```

```python
x = np.zeros((1,3,45,45))
add_info = np.zeros((1, 27))
new_data = pd.DataFrame(columns = [ 'COADD_OBJECT_ID', 'TILENAME', 'HPIX_16384', 'HPIX_4096', 'DNF_ZMEAN_SOF', 
                                     'RA', 'DEC', 'M15', 
        'Pdisk', 'Pedgeon', 'PbarGZ2', 'PbarNair10', 'Pmerg', 'Pbulge', 'Pcigar', 'TType', 'PS0', 
        'FLUX_RADIUS_G', 'FLUX_RADIUS_R', 'FLUX_RADIUS_I', 'FLUX_RADIUS_Z', 'KRON_RADIUS', 'GAP_FLUX_G',
       'MOF_BDF_FLUX_G', 'MOF_PSF_FLUX_G', 'SOF_BDF_FLUX_G', 'SOF_PSF_FLUX_G', 'MAG_AUTO_G', 'MAG_APER_4_G', 
        'MAG_APER_8_G', 'SOF_BDF_G_1', 'SOF_BDF_G_2', 'IMAFLAGS_ISO_G', 'IMAFLAGS_ISO_R', 'IMAFLAGS_ISO_I',
       'IMAFLAGS_ISO_Z'])
path_cut = 'Cutouts/'
end_name = ['_g.fits', '_i.fits', '_r.fits']

for file in tqdm(os.listdir(path_cut)):
    if(file[-6:] == 'g.fits'):
        with fits.open(path_cut + file[:-7] + end_name[0]) as hdul: cutout_g = hdul[0].data
        with fits.open(path_cut + file[:-7] + end_name[1]) as hdul: cutout_i = hdul[0].data
        with fits.open(path_cut + file[:-7] + end_name[2]) as hdul: cutout_r = hdul[0].data
            
        data_ele = data[data['COADD_OBJECT_ID'] == int(file[:-7])]
        new_data = new_data.append(data_ele.iloc[0])
        element = np.array([cutout_g, cutout_r, cutout_i])
        x = np.append(x, [element], axis = 0)
x = np.delete(x, 0, axis = 0)
new_data = new_data.astype({'TILENAME': str, 'COADD_OBJECT_ID': int, 'HPIX_16384': int, 'HPIX_4096': int})
```

```python
print(len(new_data))
new_data.head()
```

```python
table_data = Table.from_pandas(new_data)
name = 'all_better_neg_info'
primary = fits.PrimaryHDU()
image = fits.ImageHDU(x, name="IMAGE")
table = fits.BinTableHDU(data = table_data)
hdu_list = fits.HDUList([primary, image, table])
hdu_list.writeto('Cutouts/' + name + '.fits', overwrite=True)
```
