```python
import numpy as np
import healpy
import pandas as pd
```

```python
data = pd.read_csv('SDSS_catalog.csv')
```

```python
data.head()
```

```python
pixel = []
for index, row in data.iterrows():
    #if(index == 20): break
    ra, dec = row['_RA'], row['_DE']
    new_pix = healpy.ang2pix(16384, ra, dec, lonlat = True, nest = True)
    pixel.append(new_pix)
```

```python
data['HPIX_16384'] = pixel
data.head()
```

```python
data.to_csv('SDSS_catalog_hpix.csv', index=False)
```
