```python
def make_plot_all(objects, title, label, new_data):
    print(title)
    for i in range(len(objects)):
        if(i%5 == 0):
            plt.figure(figsize=(16,8))
            for j in range(5):
                if(i+j > len(objects)-1): break
                plt.subplot(1,5,j+1)
                plt.title(new_data[label].iloc[i+j])
                rgb = make_lupton_rgb(objects[i+j][2], objects[i+j][1], objects[i+j][0], Q=11., stretch=40.)
                plt.imshow(rgb, aspect='equal')
                plt.xticks([], [])
                plt.yticks([], []) 
            plt.show()
```

```python
# I need to write the DES files again (Ellipticals, Spirals and others)
def write_files(objects, data, filename):
    primary = fits.PrimaryHDU()
    image = fits.ImageHDU(objects, name="IMAGE")
    data = data.astype({'COADD_OBJECT_ID': 'int32', 'HPIX_16384': 'int32', 'HPIX_4096': 'int32'})
    data = Table.from_pandas(data)
    table = fits.BinTableHDU(data = data)
    hdu_list = fits.HDUList([primary, image, table])
    hdu_list.writeto(filename + '.fits', overwrite=True) 
```

```python
write_files(des_ellipticals, data_ellipticals, 'des_ellipticals')
write_files(des_spirals, data_spirals, 'des_spirals')
write_files(des_others, data_others, 'des_others')
```
