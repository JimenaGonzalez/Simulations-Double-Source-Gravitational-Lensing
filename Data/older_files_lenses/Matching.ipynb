```python
import numpy as np
import pandas as pd
import smatch

#If using script on terminal
#from tqdm import tqdm
from tqdm.notebook import tqdm
```

```python
data = pd.read_csv('jimena_sdss_redmagic_pixel_new.csv')
data = data.drop(labels = ['HPIX_16384', 'HPIX_4096'], axis=1)
print(len(data))
data.head()
```

```python
catalog = pd.read_csv('Y6_test.csv')
catalog = catalog.drop(labels = ['Y3_COADD_OBJECT_ID', 'RA', 'DEC', 'Z', 'VEL_DISP'], axis=1)
catalog = catalog.rename(columns={'OBJECT_Y6_RA': 'RA', 'OBJECT_Y6_DEC': 'DEC'})
print(len(catalog))
catalog.head()
```

```python
ra1 = data['RA']
dec1 = data['DEC']
ra2 = catalog['RA']
dec2 = catalog['DEC']

matches = smatch.match(ra1, dec1, 0.0002, ra2, dec2, nside=4096, maxmatch=1)
data = data.drop(labels = ['RA', 'DEC'], axis=1)
print(len(matches))
```

```python
new_data = pd.DataFrame()
for i in tqdm(range(len(matches))):
    #if(i == 10): break
    point = pd.concat( (data.iloc[matches[i][0]], catalog.iloc[matches[i][1]]), axis=0)
    new_data = new_data.append(point, ignore_index=True)
```

```python
new_data = new_data.reindex(columns=['Y6_COADD_OBJECT_ID','Y3_COADD_OBJECT_ID', 'TILENAME', 'HPIX_16384', 'HPIX_4096',
            'RA','DEC','Z', 'VEL_DISP', 'SOF_BDF_G_1', 'SOF_BDF_G_2', 'FLUX_RADIUS_G', 'FLUX_RADIUS_R',
            'FLUX_RADIUS_I', 'FLUX_RADIUS_Z', 'IMAFLAGS_ISO_G', 'IMAFLAGS_ISO_R', 'IMAFLAGS_ISO_I', 
            'IMAFLAGS_ISO_Z'])
new_data = new_data.astype({'Y6_COADD_OBJECT_ID': int, 'Y3_COADD_OBJECT_ID': int,'HPIX_16384': int, 'HPIX_4096': int})
```
