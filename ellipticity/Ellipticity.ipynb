```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import lenstronomy.Util.param_util as param
```

```python
data = pd.read_csv('theta_file.csv')
data = data.head(10000)
data
```

```python
data['my_g2'] = data['SOF_CM_G_1'] * np.tan(2*np.radians(data['THETA_J2000']))
data['my_angle'] = np.degrees(np.arctan(data['SOF_CM_G_2']/data['SOF_CM_G_1'])/2)
data['my_ratio'] = (np.cos(2*np.radians(data['THETA_J2000'])) - data['SOF_CM_G_1']) / (np.cos(2*np.radians(data['THETA_J2000'])) + data['SOF_CM_G_1'])
data
```

```python
lens_g2, lens_ratio, lens_angle = [], [], []
for index, row in data.iterrows():
    angle, ratio = param.ellipticity2phi_q(row['SOF_CM_G_1'], row['SOF_CM_G_2']) 
    g2 = (1-ratio)/(1+ratio)* np.sin(2*angle)
    lens_g2.append(g2)
    lens_ratio.append(ratio)
    lens_angle.append(np.degrees(angle))
data['lens_g2'] = lens_g2
data['lens_ratio'] = lens_ratio
data['lens_angle'] = lens_angle
#data['lens_angle_90'] = 90 + data['lens_angle']
```
