```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/jimenagonzalez/research/DSPL/Simulations-Double-Source-Gravitational-Lensing/ellipticity/shape_redshift.csv')
data = data.head(1000000)
```

```python
data = data[data['DNF_ZMEAN_SOF'] > 0.01]
data = data[data['SOF_CM_G_1'] > -5000]
data = data[data['SOF_CM_G_1'] > -5000]
data
```

```python
plt.figure(figsize=(11,15))

plt.subplot(3, 2, 1)
plt.plot(data['DNF_ZMEAN_SOF'], data['SOF_CM_G_1'], 'bo')
plt.xlabel('Redshift')
plt.ylabel('G1')
plt.xlim(0,3)

plt.subplot(3, 2, 2)
plt.plot(data['DNF_ZMEAN_SOF'], data['SOF_CM_G_2'], 'ro')
plt.xlabel('Redshift')
plt.ylabel('G2')
plt.xlim(0,3)

plt.subplot(3, 2, 3)
plt.plot(data['DNF_ZMEAN_SOF'], data['SOF_CM_G_1'], 'bo')
plt.xlabel('Redshift')
plt.ylabel('G1')
plt.xlim(0,7)

plt.subplot(3, 2, 4)
plt.plot(data['DNF_ZMEAN_SOF'], data['SOF_CM_G_2'], 'ro')
plt.xlabel('Redshift')
plt.ylabel('G2')
plt.xlim(0,7)

plt.subplot(3, 2, 5)
plt.plot(data['DNF_ZMEAN_SOF'], data['SOF_CM_G_1'], 'bo')
plt.xlabel('Redshift')
plt.ylabel('G1')

plt.subplot(3, 2, 6)
plt.plot(data['DNF_ZMEAN_SOF'], data['SOF_CM_G_2'], 'ro')
plt.xlabel('Redshift')
plt.ylabel('G2')

plt.show()
```
